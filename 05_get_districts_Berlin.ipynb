```python
import osmnx as ox
import matplotlib.pyplot as plt
```

```python
# Get all districts within Berlin by administrative level
berlin_districts = ox.features_from_place(
    "Berlin, Germany", 
    tags={"boundary": "administrative", "admin_level": "10"}
)

berlin_districts
```

```python
berlin_districts[berlin_districts['admin_level'] == '9']
```

```python
# Filter for districts and plot
berlin_districts[(berlin_districts.geometry.type == 'Polygon') & (berlin_districts['admin_level'] == '9')].plot(edgecolor="black", facecolor="none", linewidth=2, figsize=(10, 10))
plt.title("All Berlin Districts")
plt.show()
```

```python
berlin_districts[berlin_districts.geometry.type == "Point"][['geometry', 'boundary', 'admin_level', 'name', 'ref']].to_file("data/districts_point_berlin.shp")
berlin_districts[berlin_districts.geometry.type == "Polygon"][['geometry', 'boundary', 'admin_level', 'name', 'ref']].to_file("data/districts_polygon_berlin.shp")
```
